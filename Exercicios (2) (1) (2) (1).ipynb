>c) Qual é a capacidade da memória cache em bytes?

R = Isso de deve por conta do objetivo da cache, que é ser uma memória intermediária entre a memória principal e o processador, armazenando somente as informações necessárias e aumentar a velocidade das operações. Não há como ter 100% de acertos pois como a cache não armazena toda a informação, uma hora a cache terá que ir buscar dados que faltam na memória principal, ocasionando um miss.

R = No método de mapeamento direto cada bloco pertence a uma determinada linha da cache, enquanto no método associativo é preciso decidir qual bloco será retirado para dar vez ao bloco que está sendo requisitado.

```python
import math
Qblocos = 2**12
CPB = 128 # Celula por bloco
M = 8 # tamanho da celula = 1 Byte
Qlinhas = 64
largconj = 4 # Conjuntos de 4 linhas


CMP = Qblocos*CPB*M # Capacidade da MP
Qconjuntos = Qlinhas/largconj


E = math.log(CPB*Qblocos,2)

campoconjunto = math.log(Qconjuntos,2)

campocelula = math.log(CPB,2)

campoTAG = E - campocelula - campoconjunto


print("Endereço =",E,"bits")
print( )
print("TAG =",campoTAG,"bits")
print("Conjunto =",campoconjunto,"bits")
print("Célula =",campocelula,"bits")
```
